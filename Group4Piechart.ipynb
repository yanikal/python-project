The King County Births Data is a **dataset that contains information about births in King County,**

The data is specific to the year 2001 and includes a wide range of information about each birth, such as the mother's age, race, and education level, as well as the infant's birth weight, sex, and gestational age.

The data was collected and compiled by the King County Department of Public Health, which is responsible for monitoring and promoting the health and well-being of residents in the county. The purpose of collecting this data was to track trends and patterns in birth outcomes, and to identify areas where interventions and programs may be needed to improve maternal and infant health.

The data was collected through birth certificates, which are legal documents that are completed after a child is born and contain information about the birth and the parents. The data was then cleaned and organized into a format that could be used for analysis.

The King County Births Data has been used by researchers, public health professionals, and policymakers to study a wide range of topics related to maternal and infant health, such as the impact of maternal education on birth outcomes, the prevalence of preterm birth and low birth weight, and racial and ethnic disparities in birth outcomes. The data has also been used to inform public health programs and policies aimed at improving maternal and infant health in King County and beyond.






```python
import matplotlib as mpl
import matplotlib.pyplot as plt

mpl.style.use('ggplot') # optional: for ggplot-like style

# check for latest version of Matplotlib
print('Matplotlib version: ', mpl.__version__) # >= 2.0.0
```

```python
bins = np.linspace(min(df["firststep"]), max(df["firststep"]), 4)
bins
group_names = ['Low', 'Medium', 'High']
```

```python
df['firststep-binned'] = pd.cut(df['firststep'], bins, labels=group_names, include_lowest=True )
df[['firststep','firststep-binned']].head(20)
```

```python
df = df.groupby('firststep-binned', axis=0).sum()

# note: the output of the groupby method is a `groupby' object. 
# we can not use it further until we apply a function (eg .sum())
print(type(df.groupby('firststep-binned', axis=0)))

df.head()
```

```python
# autopct create %, start angle represent starting point
df['bwt'].plot(kind='pie',
                            figsize=(7, 6),
                             autopct='%1.1f%%',# add in percentages
                            startangle=90,     # start angle 90° (Africa)
                            shadow=True,       # add shadow      
                            )
plt.title('whether participation in ‘First Steps’ increases birthweights')
plt.axis('equal') # Sets the pie chart to look like a circle.
plt.legend(loc='best',bbox_transform=plt.gcf().transFigure)

plt.show()
```

**Usecase of piechart**
  

*   A pie chart is a **circular graphical representation** that is divided into slices to display numerical proportions. Each slice of the pie represents a category or segment, and its size corresponds to the percentage of that category relative to the whole.
*   Pie charts are commonly used to show the relative distribution or composition of data in a visually appealing and easy-to-understand format. They are particularly useful for displaying categorical data that is mutually exclusive and collectively exhaustive, such as market share, budget allocation, or survey results.:

```python
# autopct create %, start angle represent starting point
df['education'].plot(kind='pie',
                            figsize=(7, 6),
                            autopct='%1.1f%%', # add in percentages
                            startangle=90,     # start angle 90° (Africa)
                            shadow=False,       # add shadow      
                            )

plt.title('Percentage of education according to Race')
plt.legend(labels=df.index, loc='upper left')
plt.axis('equal') # Sets the pie chart to look like a circle.

plt.show()
```

```python
colors_list = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'lightgreen', 'pink']
#explode_list = (0.1, 0, 0, 0, 0.1, 0.1) # ratio for each continent with which to offset each wedge.
myexplode = [0.2, 0, 0, 0.1,0.1]
df['education'].plot(kind='pie',figsize=(8, 6),autopct='%1.1f%%', startangle=90,shadow=True,labels=None,pctdistance=1.12,colors=colors_list,explode=myexplode)  
# scale the title up by 12% to match pctdistance
plt.title('Percentage of education according to Race', y=1.12) 

plt.axis('equal') 

# add legend
plt.legend(labels=df.index, loc='upper left') 

plt.show()
```
