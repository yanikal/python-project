```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_capitals_in_India'
page = requests.get(url)
soup = BeautifulSoup(page.content, 'html.parser') 
# soup contains website home page data
print(page) 
# it should print 200 as a response which means we can access the website
```

```python
#printing home page data in a pretty way
print(soup.prettify())
```

```python
#Title
print(soup.title)
```

```python
print(soup.find_all('a'))
# Extract all the link/urls from anchor tag
all_links = soup.find_all('a')
for link in all_links:
    print(link.get('href'))
```

```python
#Lets print paragraph that can be found inside div
data = soup.find('div',attrs={"class":"mw-parser-output"})
print(data.p.text)
```

```python
#Now lets print all paragraphs data
data = soup.find_all('div',attrs={"class":"mw-content-ltr"})
for s in data:
    para=s.find_all('p')
    for line in para:
        print(line.text)
```

```python
#print single heading in a div
data = soup.find('span',attrs={"class":"mw-headline"})
print(data.text)
```

```python
#print all heading on home page
data = soup.find_all('div',attrs={"class":"mw-content-ltr"})
for s in data:
    heading = s.find_all('span',attrs={"class":"mw-headline"})
    for i in heading:
        print(i.text)
```

```python
#work on table data
all_tables=soup.find_all('table') 
                         # find all tables
right_table=soup.find('table', class_='wikitable sortable')
print(right_table)
# print row wise data 
for row in right_table.find_all('tr'): # get all rows
    column_marker = 0
    columns = row.find_all('td') # single row data 
    for column in columns:
        print(column.get_text())
```
